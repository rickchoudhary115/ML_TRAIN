```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=4)
X_poly=poly.fit_transform(X)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_poly,y)
```

```python
X_grid = np.arange(min(X)[0], max(X)[0], 0.1)
X_grid=X_grid.reshape(len(X_grid),1)
plt.scatter(X,y,color='red',edgecolor='black',marker="o")
plt.plot(X_grid,regressor.predict(poly.fit_transform(X_grid)),color='blue')
plt.title('Truth or Bluff (Polynomial Regressor)')
plt.xlabel('position level')
plt.ylabel('salary')
plt.tight_layout()
plt.show()
```
